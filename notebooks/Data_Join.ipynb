```python
import sqlite3
import pandas as pd
from google.colab import drive
```

```python
## this is where we join the API data with the webscraping data into
## one master dataset 'master_data.csv'

drive.mount('/content/drive')

DB_PATH = '/content/drive/My Drive/STA_141B_Project/rentals.db'

# connect to the database and read the rentals and neighborhood_data tables
conn = sqlite3.connect(DB_PATH)
df_rentals = pd.read_sql("SELECT * FROM rentals", conn)
df_neighborhoods = pd.read_sql("SELECT * FROM neighborhood_data", conn)
conn.close()

# we only want listings where a matching zip code exists
# in BOTH the rentals table and the neighborhood data
df_master = pd.merge(
    df_rentals,
    df_neighborhoods,
    left_on='zip_code',
    right_on='zip_code',
    how='inner'
)
# export to csv for use in Analysis_and_Modeling.ipnyb
df_master.to_csv('/content/drive/My Drive/STA_141B_Project/master_data.csv',
                 index=False)
```
